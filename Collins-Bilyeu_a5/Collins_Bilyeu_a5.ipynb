Task 1
Train and fine-tune a Decision Tree for the moons dataset by following these steps:

Use make_moons (n_samples=10000, noise=0.4) to generate a moons dataset

```python
#import libraries
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_moons

#create data set
X, y = make_moons(n_samples=10000, noise=0.4)
```

Use train_test_split() to split the dataset into a training set and a test set.

```python
#split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y)
```

Use grid search with cross-validation (with the help of the GridSearchCV class) to find good hyperparameter values for a DecisionTreeClassifier. (Hint: try various values for max_leaf_nodes)


```python
#create classifier class
tree_classifier = DecisionTreeClassifier()

#Use grid search with cross-validation
param_grid = [{'max_leaf_nodes': [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]}]
grid_search = GridSearchCV(tree_classifier, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

#best hyperparameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

Train it on the full training set using these hyperparameters, and measure your model's performance on the test set. You should get roughly 85% to 87% accuracy.

```python
#Fit model and evaluate test set
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)

print("Best hyperparameters:", best_params)
print("Test set accuracy:", test_accuracy)
```

Task 2
Grow a forest by following these steps:

Continuing the previous exercise, generate 1,000 subsets of the training set, each containing 100 instances selected randomly. (Hint: you can use Scikit-Learnâ€™s ShuffleSplit class for this.)


```python
from sklearn.model_selection import ShuffleSplit

shuffle_split = ShuffleSplit(n_splits=1000, train_size = 100)
```

Train one Decision Tree on each subset, using the best hyperparameter values found in the previous exercise. Evaluate these 1,000 Decision Trees on the test set. Since they were trained on smaller sets, these Decision Trees will likely perform worse than the first Decision Tree.


```python
predictions = []

for train_index, _ in shuffle_split.split(X_train):
    #subsets
    X_subset, y_subset = X_train[train_index], y_train[train_index]

    #Train the decision tree
    best_model.fit(X_subset, y_subset)

    # Predict on the test set and store predictions
    y_pred = best_model.predict(X_test)
    predictions.append(y_pred)

for prediction in predictions:
    print("accuracy: ", accuracy_score(y_test, prediction))
```

Now comes the magic. For each test set instance, generate the predictions of the 1,000 Decision Trees, and keep only the most frequent
prediction (you can use SciPyâ€™s mode() function for this). This approach gives you majority-vote predictions over the test set.

```python
from scipy.stats import mode

predictions = np.array(predictions)

#Take mode of each test value set
final_predictions = mode(predictions, axis=0).mode
```

Evaluate these predictions on the test set: you should obtain a slightly higher accuracy than your first model (about 0.5 to 1.5% higher). Congratulations, you have trained a Random Forest classifier!


```python
#Calculate accuracy of the final predictions
final_accuracy = accuracy_score(y_test, final_predictions)

print("Final accuracy:", final_accuracy)
```
